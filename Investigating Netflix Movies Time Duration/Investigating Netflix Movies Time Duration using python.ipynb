```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load the CSV
netflix_df = pd.read_csv('netflix_data.csv', index_col= 0)
print(netflix_df.iloc[:10])
#Removing the TV shows from data
for lab, row in netflix_df.iterrows():
    row = netflix_df[netflix_df['type'] == 'Movie']
# store as netflix_subset and preview the new data frame
netflix_subset = row
print(netflix_subset.iloc[:])
```

```python
#keeping only the columns "title", "country", "genre", "release_year", "duration", and saving this into a new DataFrame called netflix_movies.

netflix_movies = netflix_subset[["title", "country","genre","release_year","duration"]]
print(netflix_movies.iloc[:,3:5])

#  movies that are shorter than 60 minutes, saving the resulting DataFrame as short_movies

short_movies = netflix_movies[netflix_movies['duration']<60]
print(short_movies.iloc[0:20,[0,3,4]])
```

```python
# assign colors to each genre group by using the empty dectionary and if statments

colors= []

for lab, row in netflix_movies.iterrows():
    if row['genre'] == 'Documentaries':
        colors.append('Grey')
    elif row ['genre'] == 'Dramas':
        colors.append('Brown')
    elif row ['genre'] == 'Stand-Up':
        colors.append('Yellow')
    elif row ['genre'] == 'Children':
        colors.append('Cyan')
    elif row ['genre'] == 'Comedies':
        colors.append('Green')
    else :
        colors.append('Black')
        
colors[:20]
```

```python
fig = plt.figure(figsize=(12, 8))
plt.scatter(netflix_movies['release_year'], netflix_movies['duration'], c=colors)
plt.xlabel('Release year')
plt.ylabel('Duration (min)')
plt.title("Movie Duration by Year of Release")
```
